```python
import pandas as pd
```

```python
#reading our main dataset
housing_data=pd.read_csv("Datasets/Melbourne_housing_FULL.csv")
housing_data.head()
```

```python
#get the list of columns titles
housing_data.columns
```

```python
#Number of rows and columns
housing_data.shape
```

```python
#check number of values for each title
housing_data.count()
```

```python
# droo NAN values from the dataset and clean our dataset
housing_cleaned_data=housing_data.dropna()
housing_cleaned_data.count()
```

```python
#check types of values 
housing_cleaned_data.dtypes
```

```python
#convert postcode type from float to ineteger and then object for getting it ready to merge. 
convert_types={"Postcode" : int}
housing_cleaned_data=housing_cleaned_data.astype(convert_types)
convert_types={"Postcode" : object , "Price": int , "Propertycount" : int ,
               "Bedroom2" : int, "Bathroom" : int , "Car" : int , "YearBuilt" : int , "Landsize" : int,
               "BuildingArea": int, "Propertycount" : int}
housing_cleaned_data=housing_cleaned_data.astype(convert_types)
```

```python
housing_cleaned_data.head()
```

```python
#save housing_data to csv file
housing_cleaned_data.to_csv("Datasets/cleaned_housing_data.csv", encoding="utf-8", index=False)
```

```python
#import Census data
census_data=pd.read_csv("Datasets/VIC/2021Census_G02_VIC_POA.csv")
census_data.head()
```

```python
census_data.columns
```

```python
#change postcode column title and type and removing POA from the postcode
census_data["POA_CODE_2021"]=census_data["POA_CODE_2021"].str.replace('POA', '')
census_data=census_data.rename(columns={"POA_CODE_2021": "Postcode" })
convert_types={"Postcode" : object}
census_data=census_data.astype(convert_types)
census_data.dtypes
```

```python
#save cleaned census dataset to csv file
census_data.to_csv("Datasets/census.csv", encoding="utf-8", index=False)
```
