```python
import pandas as pd
```

```python
#reading housing data and census 
census_data=pd.read_csv("census.csv")
housing_data=pd.read_csv("cleaned_housing_data.csv")
```

```python
census_data.head()
```

```python
# merge both datasets based on postcode column
merged_df=pd.merge(housing_data,census_data, on="Postcode" , how="inner")
```

```python
merged_df.head()
```

```python
merged_df.count()
```

```python
#save the merged dataset
merged_df.to_csv("Merged.csv")
```

```python
#check the columns title
merged_df.columns
```

```python
#rename some of columns titles
merged_df=merged_df.rename(columns={"Bedroom2": "Number of Bedroom",
                                    "Bathroom":"Number of Bathroom",
                                    "SellerG":"Real Estate Agent",
                                    "Car" : "number of carpark"})
```
